```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importar el separador de muestras para entrenamiento y pruebas
from sklearn.model_selection import train_test_split
# Importar el clasificador: Árbol de decisión
from sklearn.tree import DecisionTreeClassifier, plot_tree
# Importar las métricas para medir la eficiencia del modelo
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
```

```python
train, test = train_test_split(data, test_size=0.4, stratify=data['species'], random_state=42)
```

```python
fn = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
cn = ['setosa', 'versicolor', 'virginica']
```

```python
# Conjunto de pruebas siguiendo el modelo y = mx + b
x_test = train[fn]
y_test = train['species']
```

```python
# Árbol de decisión
mod_dt = DecisionTreeClassifier(max_depth=3, random_state=112)
# Ajustar el modelo
mod_dt.fit(x_test, y_test)
# Probar el modelo
prediccion = mod_dt.predict(x_test)
```

```python
# Importancia de los predictores
mod_dt.feature_importances_
```

```python
# Visualizar las reglas de clasificación
plt.figure(figsize=(10,8))
plot_tree(mod_dt,feature_names=fn,class_names=cn,filled=True);
```

```python
# Eficiencia del modelo
eficiencia = metrics.accuracy_score(y_test, prediccion)
eficiencia
```

```python
coso = pd.DataFrame(x_test)
coso['species'] = y_test
coso['prediccion'] = prediccion
```
