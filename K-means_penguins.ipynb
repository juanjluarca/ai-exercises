## Clasificador K-Means (Clustering - No supervisado)
### Objetivo: Identificar la especie de pingüino (Adelie, Chinstrap, Gentoo) en base a: largo y profundidad de pico, largo de aleta y masa corporal

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.info()
```

```python
df.rename({
    'species': 'Especie',
    'island': 'Isla',
    'bill_length_mm': 'Largo_Pico',
    'bill_depth_mm': 'Profundidad_Pico',
    'flipper_length_mm': 'Largo_Aleta',
    'body_mass_g': 'Masa_Corporal',
    'sex': 'Sexo'
}, axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

## Análisis exploratorio

```python
df.describe().T
```

```python
df.Especie.value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
sns.set()
```

```python
df.Largo_Pico.plot.hist()
plt.title('Distribución de Largo de Pico')
plt.xlabel('Largo de Pico (mm)');
```

```python
df.groupby('Especie')[['Largo_Pico', 'Profundidad_Pico', 'Largo_Aleta', 'Masa_Corporal']].describe()
```

```python
df.plot.scatter(x='Largo_Pico', y='Largo_Aleta')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
col_escalar = ['Largo_Pico', 'Profundidad_Pico', 'Largo_Aleta', 'Masa_Corporal']
datos_escalados = df.copy()
datos_escalados[col_escalar] = scaler.fit_transform(df[col_escalar])
```

```python
df
```

```python
datos_escalados
```

```python
datos_escalados.plot.scatter(x='Largo_Pico', y='Largo_Aleta')
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=3, random_state=16)
modelo.fit(datos_escalados[col_escalar])
```

```python
datos_escalados['Segmento K'] = modelo.predict(datos_escalados[col_escalar])
datos_escalados
```

```python
datos_escalados['Segmento K'].value_counts()
```

```python
marcador = ['x', '*', '.']
```

```python
for segmento in range(3):
    temporal = datos_escalados[datos_escalados['Segmento K']==segmento]
    plt.scatter(temporal.Largo_Pico, temporal.Largo_Aleta, marker=marcador[segmento], label='Segmento ' + str(segmento))
    plt.xlabel('Largo de Pico (estandarizado)')
    plt.ylabel('Largo de Aleta (estandarizado)')
    plt.legend();
```
