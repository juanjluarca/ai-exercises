```python
from transformers import BertTokenizer
```

```python
import torch
torch.cuda.empty_cache()
```

```python
tokenizer = BertTokenizer.from_pretrained('google-bert/bert-base-uncased')
```

```python
texts_en = [
  "This product exceeded my expectations and works perfectly.",
  "I am very disappointed because the quality is much lower than advertised.",
  "The item arrived on time and performs exactly as described.",
  "It stopped working after a week, so I would not recommend it.",
  "Great value for the price, I would definitely buy it again.",
  "The design looks nice, but the performance is very poor.",
  "I am extremely satisfied with this purchase and use it every day.",
  "Customer support was unhelpful and the product feels cheap.",
  "It does its job well and feels sturdy and reliable.",
  "The product is overpriced and does not meet basic expectations."
]
```

```python
texts_es = [
  "Este producto superó mis expectativas y funciona perfectamente.",
  "Estoy muy decepcionado porque la calidad es mucho más baja de lo anunciado.",
  "El artículo llegó a tiempo y funciona exactamente como se describe.",
  "Dejó de funcionar después de una semana, así que no lo recomendaría.",
  "Excelente relación calidad-precio, definitivamente lo compraría de nuevo.",
  "El diseño se ve bien, pero el rendimiento es muy pobre.",
  "Estoy extremadamente satisfecho con esta compra y lo uso todos los días.",
  "El servicio de atención al cliente no fue útil y el producto se siente barato.",
  "Cumple bien su función y se siente resistente y confiable.",
  "El producto es caro y no cumple con las expectativas básicas."
]
```

```python
tokens_es = tokenizer.tokenize(texts_es[0])
tokens_en = tokenizer.tokenize(texts_en[0])
```

```python
tokens_es
```

```python
tokens_en
```

```python
tokens_id_en = tokenizer.convert_tokens_to_ids(tokens_en)
tokens_id_en
```

```python
texto_en_decode = tokenizer.decode(tokens_id_en)
texto_en_decode
```

```python
for i, oracion in enumerate(texts_en):
    # encode añade automáticamente [CLS] y [SEP]
    codificado = tokenizer._encode_plus(
        oracion,
        add_special_tokens=True,
        max_length=20,
        padding='max_length',
        truncation=True,
        return_tensors='pt',
        return_attention_mask=True
    )
    print(f'\n Oración {i+1}: {oracion}')
    print(f'IDs: {codificado["input_ids"][0].tolist()}')
    print(f'Máscara de atención: {codificado["attention_mask"][0].tolist()}')
```

```python
batch_encode = tokenizer(
    texts_en,
    max_length=20,
    padding='max_length',
    truncation=True,
    return_tensors='pt',
)
```
